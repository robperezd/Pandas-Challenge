# Executive Summary

Based on the information provided and the period given, Heroes Of Pymoli had registered a total number of 780 players and had generated a total revenue of $2,380 USD.
Audience: 84percent of their players are Male.
Age Demographics: The group between 20-24 years represent 47% of Total Players and 47% of Total Revenue, follow by the group between 15-19 years that represent 17.4% of Total Players and 17.3% of Total Revenue and finally the group between 25-29 years represent 13% of Total Players and 12.3% of Revenue. Together these 3 groups represent almost 80% of Total Players and Revenue.

## Purchasing Analysis (Age)

```python
df=pd.DataFrame(purchase_data)
df
```

```python
bins = [0,10,14,19,24,29,34,39,45]

group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

df["Age Ranges"] = pd.cut(df["Age"], bins, labels=group_names, include_lowest=True)
df
```

```python
Purchasing_D = df.groupby(["Age Ranges"])
Purchasing_DF = (Purchasing_D["Purchase ID"]).count()
Purchasing_DF
```

```python
Purchasing_D = df.groupby(["Age Ranges"])
Purchasing_DF_Avg_Pur = round( (Purchasing_D["Price"]).mean(),2)
Purchasing_DF_Avg_Pur
```

```python
Purchasing_D = df.groupby(["Age Ranges"])
Purchasing_DF_Tot = round((Purchasing_D["Price"]).sum(),2)
Purchasing_DF_Tot
```

```python
Purchasing_D = df.groupby(["Age Ranges"])
Purchasing_DF_Avg_Tot_Pur = round( (Purchasing_D["Price"]).max(),2)
Purchasing_DF_Avg_Tot_Pur
```

```python
Purchasing_Analysis_Summary_df = pd.DataFrame({"Purchase Count": Purchasing_DF,
                                      "Average Purchase Price": Purchasing_DF_Avg_Pur,
                                      "Total Purchase Value": Purchasing_DF_Tot,
                                      "Avg Total Purchase per Person":Purchasing_DF_Avg_Tot_Pur         
                                 })
Purchasing_Analysis_Summary_df
```
